```python
import pandas as pd
import numpy as np
```

```python
initial_data = pd.read_csv('data.csv')
PRIZ=10
SIZ=2
```

```python
initial_data.head()
```

```python
initial_data.drop(initial_data.columns[0], axis=1, inplace=True)
```

```python
for column in initial_data.columns:
    if "mean" not in column:
        if 'diagnosis' not in column:
            initial_data.drop(column, axis = 1, inplace=True)
```

```python
initial_data['diagnosis']=initial_data['diagnosis'].map({'M':1,'B':0})
```

```python
rez = initial_data.mean().values
```

```python
for i,column in enumerate(initial_data.columns):
        r=rez[i]
        initial_data[column]=initial_data[column].apply(lambda x: 1 if x >r else 0)
```

```python
initial_data
```

```python
dfM=initial_data[initial_data['diagnosis']==1]
dfM.drop('diagnosis', axis = 1, inplace=True)
dfB=initial_data[initial_data['diagnosis']==0]
dfB.drop('diagnosis', axis = 1, inplace=True)
```

```python
m_m=len(dfM)
```

```python
m_b=len(dfB)
```

```python
B=np.zeros( (SIZ, PRIZ) )
```

```python
def countB(N,M,df):
    for i in range(PRIZ):
        B[N][i]=sumX(M,df.values,i)/M        
def sumX(M,x,t):
    s=0
    for j in range(M):
        s+=x[j][t]
    return s    
        
countB(0,m_m,dfM)        
countB(1,m_b,dfB)    
```

```python
BT=np.zeros(PRIZ)
A=np.zeros( (SIZ, PRIZ) )
```

```python
for i in range(PRIZ):
    s=0
    for j in range(SIZ):
        s+=B[j][i]
    BT[i]=s/SIZ
for i in range(PRIZ):
    for j in range(SIZ):
        A[j][i]=abs(B[j][i]-BT[i])  
```

```python
mat = np.matrix(A)
with open('outfile.txt','wb') as f:
    for line in mat:
        np.savetxt(f, line, fmt='%.5f')
```
